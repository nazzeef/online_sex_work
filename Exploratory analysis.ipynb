#### 4.1.5 Time spent chatting

We will replace data in this column with the total number of minutes indicated. 

Some entries have the `H:M` (hour:minute) format, while others have what appears to be the `D:H:M` (day:hour:minute) format.

We will define a `get_n_minutes` function in order to replace the data into total number of minutes. We will then convert this to an integer. 

```python
def get_n_minutes(row):
    time_components = row.split(':')
    if len(time_components) == 2:
        return int(time_components[0]) * 60 + int(time_components[1])
    elif len(time_components) == 3:
        return int(time_components[0]) * 1440 + int(time_components[1]) * 60 + int(time_components[2])
```

```python
df['Time_spent_chating_H:M'] = df['Time_spent_chating_H:M'].str.replace(' ', '')
df['Time_spent_chating_H:M'] = df['Time_spent_chating_H:M'].apply(get_n_minutes)
```

#### 4.2.1 Gender (contains missing data)

In the Gender column, we fill some missing values using some simple conditions (if the entry is, for example, homosexual, and looking for men, we fill that entry with male), using the fill_gender_na function below. Then in every entry, we change the data to whether it specifies female or not.

```python
df['Gender'].value_counts(dropna=False)
```

```python
def fill_gender_na(row):
    if row['Sexual_orientation'] == 'Homosexual':
        if row['Looking_for'] == 'Men':
            return 'male'
        elif row['Looking_for'] == 'Women':
            return 'female'
    elif row['Sexual_orientation'] == 'Heterosexual':
        if row['Looking_for'] == 'Men':
            return 'female'
        elif row['Looking_for'] == 'Women':
            return 'male'
    
    return np.nan
```

```python
fill_values = df.apply(fill_gender_na, axis=1)
df['Gender'].fillna(fill_values, inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

However, they do contain some missing data. Also, under the Looking_for column, there are respondents who responded with 'Nobody' or 'Nobody_but_maybe'. We will change these values to either Men, Women or Men_and_Women. 

We can impute the values for these columns by cross-referencing them against one another.

```python
def resetting_looking_for(row):
    if row['Sexual_orientation'] == 'Homosexual':
        if row['Gender'] == 'female':
            row['Looking_for'] = 'Women'
        if row['Gender'] == 'male':
            row['Looking_for'] = 'Men'
    if row['Sexual_orientation'] == 'Heterosexual':
        if row['Gender'] == 'female':
            row['Looking_for'] = 'Men'
        if row['Gender'] == 'male':
            row['Looking_for'] = 'Women'
    if row['Sexual_orientation'] == 'bicurious' or row['Sexual_orientation'] == 'bisexual':
        row['Looking_for'] = 'Men_and_Women'
    return row
```

### 4.3 Change 'Member since' to datetime

We will change Member since from a string to a datetime. 

An entry was evidently inputted incorrectly (it contains 0,278159722), hence, it was replaced using the **mode( )** function.

Some entries contain dnes, which translates to "today" in Czech (probably where the dataset originates from), so it was arbitrarily replaced with the maximum date among the other entries, with 1 more day added thus assumed to be the date when the data was extracted (i.e the date of "today"). 

We will then create new columns for representing the Day, Month and Year. 

```python
df.Member_since.unique()
```

```python
df['Member_since'][19410]
```

```python
df['Member_since'].replace(to_replace='0,278159722', value=df['Member_since'].mode()[0], inplace=True)
df['Member_since'].replace(to_replace='dnes', value=np.nan, inplace=True)
df['Member_since'] = pd.to_datetime(df['Member_since'], format='%d.%m.%Y')
```

```python
from datetime import timedelta
```

```python
df['Member_since'] = df['Member_since'].fillna(df['Member_since'].max() + timedelta(days=1))
df['Member_since'].head()
```

```python
# Creating new columns for Year, Month, and Day
df.insert(9,'Member_since_year',df.Member_since.dt.year)
df.insert(10,'Member_since_month',df.Member_since.dt.month)
df.insert(11,'Member_since_day',df.Member_since.dt.day)
```

### 5.1 Growth in membership of a European adult forum from 2009 to 2017 segregated by users assessed to be High-risk and No-risk

## 5.5 Network of friends

```python
import networkx as nx
import matplotlib.patches as mpatches
```

```python
# Dropping the data values from the data set where the Risk is unknown_risk:
df.drop(df[(df.Risk != 'High_risk') & (df.Risk != 'No_risk')].index, inplace = True)
```

```python
df['Friends_ID_list'] = df['Friends_ID_list'].replace('nan', np.nan)
```

```python
df_network = df[df['Friends_ID_list'].isnull() == False]['Friends_ID_list']
```

```python
graph = nx.Graph()
graph.add_nodes_from(list(df_network.index))
```

```python
for id_network in df_network.index:
    friend_ids = list(map(int, df_network.loc[id_network].split(',')))
    for friend_id in friend_ids:
        graph.add_edge(id_network, friend_id)
```

```python
df_norisk = df[df['Risk'] == 'No_risk']

df_risk = df[df['Risk'] == 'High_risk']
```

```python
f, ax = plt.subplots(1, 1, figsize=(20, 10))

pos = nx.spring_layout(graph)


norisk_nodelist = list(df_norisk[df_norisk['Friends_ID_list'].isnull() == False].index)
risk_nodelist = list(df_risk[df_risk['Friends_ID_list'].isnull() == False].index)

# Drawing the network for no-risk users:
nx.draw_networkx_nodes(
    graph,
    pos,
    nodelist = norisk_nodelist,
    node_color = 'b',
    node_size = 50
)

# Drawing the network for high-risk users:
nx.draw_networkx_nodes(
    graph,
    pos,
    nodelist = risk_nodelist,
    node_color = 'r',
    node_size = 50
)

labels = {}
for node in norisk_nodelist: labels[node] = node
for node in risk_nodelist: labels[node] = node
nx.draw_networkx_labels(graph, pos, labels, font_size=10)

nx.draw_networkx_edges(graph, pos, edge_color='y')

patches = [
    mpatches.Patch(color='b', label='No Risk'),
    mpatches.Patch(color='r', label='High Risk')
]
plt.legend(handles=patches)

plt.title('Network of friends between users assessed to be High risk and No risk', fontsize = 20)

plt.show()
```
